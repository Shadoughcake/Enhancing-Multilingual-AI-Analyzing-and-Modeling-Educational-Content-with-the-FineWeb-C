```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_parquet("hf://datasets/data-is-better-together/fineweb-c/dan_Latn/train-00000-of-00001.parquet")
# REMOVE PROBLEMATIC LABELS FROM DATASET
df = df[df['problematic_content_label_present'] == False]
```

```python
sort_order = {
    "None": 0,
    "Minimal": 1,
    "Basic": 2,
    "Good": 3,
    "Excellent": 4
}

def encoding(edu_list):
    labels = set(edu_list)
    L = [0 for _ in range(len(sort_order))]
    for label in labels:
        if label in sort_order:
            L[sort_order[label]] = 1
    return L

def disagreement(encoding):
    indices = [i for i, val in enumerate(encoding) if val == 1]
    print(len(indices), indices)
    if len(indices) > 2:
        return True
    
    if len(indices) == 2:
        dist = abs(indices[0] - indices[1])
        print(dist)
        if dist > 1:
            return True
    
    return False


df["encoding"] = df["educational_value_labels"].apply(encoding)
df["disagreement"] = df["encoding"].apply(disagreement)

hd_df = df[df["disagreement"] == True]
```

# Process Final Fineweb-C

```python
df = pd.read_parquet("hf://datasets/data-is-better-together/fineweb-c/dan_Latn/train-00000-of-00001.parquet") # Fineweb-C

A_csv = pd.read_csv('fineweb-c_relabled_aleks.csv', index_col=0)
W_csv = pd.read_csv('Will.csv', index_col=0)

relabeled_df = pd.concat([A_csv, W_csv], ignore_index=False)
```

```python
def update_labels(Df1,Df2):
    Df1 = Df1.set_index('id')
    Df2 = Df2.set_index('id')

    Df1.update(Df2[['educational_value_labels']])

    return Df1.reset_index()
```

```python
updated_df = update_labels(df, relabeled_df)
```

```python
hd_samples = hd_df.sample(5, random_state=42)
display(hd_samples[['id', 'text', 'educational_value_labels']])


hd_samples_ids = hd_samples['id'].tolist()

updated_samples = updated_df[updated_df['id'].isin(hd_samples_ids)]
display(updated_samples[['id', 'text', 'educational_value_labels']])
```

```python
updated_df.to_csv('fineweb-c_relabled.csv', index=True)
```
