```python
import pandas as pd
import numpy as np
import ast
```

```python
# df = pd.read_parquet("hf://datasets/data-is-better-together/fineweb-c/dan_Latn/train-00000-of-00001.parquet")
df = pd.read_csv("fineweb-c_relabled.csv")
df["educational_value_labels"] = df["educational_value_labels"].apply(ast.literal_eval)
# REMOVE PROBLEMATIC LABELS FROM DATASET
# df = df[df['problematic_content_label_present'] == False]
```

```python
an1 = pd.read_csv("annotations/annotations.csv")
an2 = pd.read_csv("annotations/annotations-2.csv")
an3 = pd.read_csv("annotations/annotations-3.csv")

an1['label'] = an1['label'].apply(ast.literal_eval)
an2['label'] = an2['label'].apply(ast.literal_eval)
an3['label'] = an3['label'].apply(ast.literal_eval)
```

```python
annotated_combined = pd.DataFrame()

# For each row in all an's:
for i in range(len(an1)):
    # Merge all labels into one df:
    an_anotatins = [an1.iloc[i]['label'][0], an2.iloc[i]['label'][0], an3.iloc[i]['label'][0]]
    fbc_annotations = ast.literal_eval(df[df["Unnamed: 0"] == int(an1.iloc[i]["id"])]["educational_value_labels"].values[0])
    for j in fbc_annotations:
        an_anotatins.append(j)

    combined_row = pd.DataFrame({
        'id': an1.iloc[i]['id'],
        'label': [an_anotatins]
    })
    annotated_combined = pd.concat([annotated_combined, combined_row], ignore_index=False)

annotated_combined.rename(columns={'label': 'education_value_labels'}, inplace=True)
annotated_combined.sample(5)
```

```python
sort_order = {
    "None": 0,
    "Minimal": 1,
    "Basic": 2,
    "Good": 3,
    "Excellent": 4
}

def encoding(edu_list):
    labels = set(edu_list)
    L = [0 for _ in range(len(sort_order))]
    for label in labels:
        if label in sort_order:
            L[sort_order[label]] = 1
    return L

def disagreement(encoding):
    indices = [i for i, val in enumerate(encoding) if val == 1]
    print(len(indices), indices)
    if len(indices) > 2:
        return True
    
    if len(indices) == 2:
        dist = abs(indices[0] - indices[1])
        print(dist)
        if dist > 1:
            return True
    
    return False


df["encoding"] = df["educational_value_labels"].apply(encoding)
df["disagreement"] = df["encoding"].apply(disagreement)

hd_df = df[df["disagreement"] == True]
```

```python
annotated_combined['encoding'] = annotated_combined['education_value_labels'].apply(encoding)
annotated_combined['disagreement'] = annotated_combined['encoding'].apply(disagreement)
hd_annotated_combined = annotated_combined[annotated_combined["disagreement"] == True]
```

```python
# df = pd.read_parquet("hf://datasets/data-is-better-together/fineweb-c/dan_Latn/train-00000-of-00001.parquet") # Fineweb-C
df = pd.read_csv("fineweb-c_relabled.csv")

A_csv = pd.read_csv('fineweb-c_relabled_aleks.csv', index_col=0)
W_csv = pd.read_csv('Will.csv', index_col=0)

# relabeled_df = pd.concat([A_csv, W_csv], ignore_index=False)

relabeled_df = pd.read_csv('annotations_relabled.csv', index_col=0)
```

```python
def update_labels(Df1,Df2):
    Df1 = Df1.set_index('id')
    Df2 = Df2.set_index('id')

    Df1.update(Df2[['educational_value_labels']])

    return Df1.reset_index()
```

```python
updated_df = update_labels(df, relabeled_df)
```

```python
hd_samples = hd_df.sample(5, random_state=42)
display(hd_samples[['id', 'text', 'educational_value_labels']])


hd_samples_ids = hd_samples['id'].tolist()

updated_samples = updated_df[updated_df['id'].isin(hd_samples_ids)]
display(updated_samples[['id', 'text', 'educational_value_labels']])
```

```python
updated_df['educational_value_labels'] = updated_df['educational_value_labels'].apply(lambda x: list(x) if isinstance(x, np.ndarray) else x)
updated_df.to_csv('fineweb-c_relabled.csv', index=True)
```
